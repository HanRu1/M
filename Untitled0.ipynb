```python
from metrics import load
from metrics import metrics_all

label, la_header = load(image1_path)
predict, pr_header = load(image2_path)

def evaluate_images(predict, label, metric_name):
    function_mapping = {
        'accuracy': metrics_all.accuracy,
        'precision': metrics_all.precision,
        'recall': metrics_all.recall,
        'specificity': metrics_all.specificity,
        'f1_score': metrics_all.f1_score,
        'dice': metrics_all.dice_coefficient,
        'iou': metrics_all.iou,
        'g_mean': metrics_all.g_mean,
        'mae': metrics_all.mae,
        'hd': metrics_all.hausdorff_distance,
        '95hd': metrics_all.hausdorff_95,
        'ssim': metrics_all.calculate_ssim,
        'ncc': metrics_all.ncc,
        'psnr': metrics_all.calculate_psnr,
        'kappa': metrics_all.cohen_kappa,
        'log_loss': metrics_all.log_loss,
        'fpr': metrics_all.fpr,
        'fnr': metrics_all.fnr,
        'voe': metrics_all.voe,
        'rvd': metrics_all.rvd,
        'sensitivity': metrics_all.sensitivity,
        'jaccard': metrics_all.jaccard_coefficient,
        'tnr': metrics_all.tnr,
        'tpr': metrics_all.tpr,
        'confusion_matrix': metrics_all.calculate_confusion_matrix,
        'roc': metrics_all.roc,
        'auc': metrics_all.auc,
        'error_rate': metrics_all.error_rate,
        'mcc': metrics_all.mcc,
        'fdr': metrics_all.fdr,
        'npv': metrics_all.npv,
        'balanced_accuracy': metrics_all.balanced_accuracy,
        'mse': metrics_all.mse,
        'MI': metrics_all.mutual_information,
        'NMI': metrics_all.normalized_mutual_information,
        'CC': metrics_all.correlation_coefficient,
        'cross_entropy': metrics_all.cross_entropy,
    }
    return function_mapping[metric_name](predict, label, int(target_value))

while(True):
  metric_name = input("请输入指标(exit结束)：")
  if metric_name == 'exit':
    break
  elif metric_name == "fid":
    value = metrics_all.calculate_fid(predict, label)
  else:
    value = evaluate_images(predict, label, metric_name)

  print(metric_name,":",value,"\n")
```
